```python
def random_rows_after_2000(data_frame, num_rows):
    total_rows = data_frame.shape[0]
    eligible_indices = list(range(2001, total_rows))

    selected_indices = random.sample(eligible_indices, num_rows)
    selected_rows = data_frame.iloc[selected_indices]

    return selected_rows
```

```python
class NeuralNet(object): 
    def __init__(self, input_size): 
        # Generate random numbers 
        np.random.seed(1) 
  
        # Assign random weights to a 3 x 1 matrix, 
        self.synaptic_weights = 2 * np.random.rand(input_size, 1) - 1
        # self.synaptic_weights = 2 * np.random.rand(1, input_size) - 1
  
    # The Sigmoid function 
    def __sigmoid(self, x): 
        return 1 / (1 + np.exp(-x)) 
  
    # The derivative of the Sigmoid function. 
    # This is the gradient of the Sigmoid curve. 
    def __sigmoid_derivative(self, x): 
        return x * (1 - x) 
  
    # Train the neural network and adjust the weights each time. 
    def train(self, inputs, outputs, training_iterations): 
        for iteration in range(0, training_iterations): 
            # Pass the training set through the network. 
            output = self.learn(inputs) 
  
            # Calculate the error 
            error = outputs - output 
  
            # Adjust the weights by a factor 
            factor = np.dot(inputs.T, error * self.__sigmoid_derivative(output))
            factor = factor.mean(axis=1, keepdims=True)  # Compute the mean along axis 1
            self.synaptic_weights += factor
  
        # The neural network thinks. 
  
    def learn(self, inputs): 
        return self.__sigmoid(np.dot(inputs, self.synaptic_weights)) 
```
